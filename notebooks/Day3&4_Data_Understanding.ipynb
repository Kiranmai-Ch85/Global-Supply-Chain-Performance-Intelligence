**DATA VALIDATION**

Detect data-type issues (numeric-as-text, malformed dates in InvoiceDate, PODate

```python
datasets = {
    "vendor_invoice": vendor_invoice,
    "purchase_prices": purchase_prices,
    "vendor_sales_summary": vendor_sales_summary,
    "purchases": purchases,
    "begin_inventory": begin_inventory,
    "sales": sales,
    "end_inventory": end_inventory,
    "procurement_kpi": procurement_kpi,
    "purchase_orders": purchase_orders,
    "supply_chain": supply_chain
}

date_columns = [
    "InvoiceDate",
    "PODate",
    "ReceivingDate",
    "PayDate",
    "SalesDate",
    "startDate",
    "endDate",
    "Order_Date",
    "Delivery_Date",
    "order_date",
    "promised_date",
    "receipt_date"
]

for name, df in datasets.items():

    print("="*70)
    print(name.upper())
    print("="*70)

    for col in date_columns:

        if col in df.columns:

            invalid_dates = pd.to_datetime(df[col], errors="coerce").isna().sum()

            print(f"{col} : {invalid_dates}")

    print()
```

Detect Invalid Values (Quantity, Price, Volume)

```python
numeric_columns = [
    "Quantity",
    "PurchasePrice",
    "Price",
    "SalesPrice",
    "ActualPrice",
    "Unit_Price",
    "Negotiated_Price",
    "Volume"
]

for name, df in datasets.items():

    print("="*70)
    print(name.upper())
    print("="*70)

    for col in numeric_columns:

        if col in df.columns:

            temp = pd.to_numeric(df[col], errors="coerce")

            negative = (temp < 0).sum()

            zero = (temp == 0).sum()

            print(f"{col}")
            print(f"Negative Values : {negative}")
            print(f"Zero Values     : {zero}")
            print()

    print()
```

Detect Inconsistent Categories (Classification, City, Item_Category)

```python
category_columns = [
    "Classification",
    "City",
    "Item_Category"
]

for name, df in datasets.items():

    print("="*70)
    print(name.upper())
    print("="*70)

    for col in category_columns:

        if col in df.columns:

            print(f"\nColumn : {col}")
            print("Unique Values :", df[col].nunique(dropna=False))
            print(df[col].value_counts(dropna=False))

    print()
```
